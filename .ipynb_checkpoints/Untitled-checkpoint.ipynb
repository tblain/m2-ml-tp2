```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

# Partie I

```python
credit = pd.read_csv("credit_scoring.csv", sep=";", header=0)
```

```python
credit
```

```python
X = credit.drop("Status", axis=1)
y = credit.Status
```

```python
mp = 100 * np.sum(y==0) / len(y)
bp = 100 * np.sum(y==1) / len(y)

print("% Bons payeurs : {0:.2f} % mauvais payeurs : {1:.2f}".format(bp, mp))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

DT = DecisionTreeClassifier(criterion="gini", random_state=1) # CARTE
DT.fit(X_train, y_train)
pred = DT.predict(X_test)
```
